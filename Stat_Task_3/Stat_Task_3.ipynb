##### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
X = np.array([100,80,75,77,89,33,45,25,65,17,30,24,57,55,70,75,65,84,90,150])
x_ = X.sum()/len(X)

s = 0
for i in range (len(X)):
    s += abs((X[i] - x_))**2
    
S = s/(len(X))
S1 = sqrt(s/(len(X)-1))
S2 = sqrt(s/(len(X)))

print(f'Среднее арифметическое: {x_}')
print(f'Среднее квадратичное отклонение: {S}')
print(f'Смещенная оценка дисперсии: {S1}')
print(f'Несмещенная оценка дисперсии: {S2}')
```

```python
print(np.mean(X))
print(np.std(X, ddof=1))
print(np.std(X))
print(np.var(X))
```

##### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?


```python
def comb(n,k):
    return f(n)/(f(k)*f(n-k))
```

```python
p1 = comb(5,2) / comb(8,2) * (comb(5,1) * comb(7,3) / comb(12,4)) # (2:1)
p2 = (comb(5,1) * comb(3,1) / comb(8,2)) * (comb(5,2) * comb(7,2) / comb(12,4)) # (1:2)
p3 = comb(3,2) / comb(8,2) * (comb(5,3) * comb(7,1) / comb(12,4)) # (0:3)
p = p1 + p2 + p3
print(f'p1: {p1}\np2: {p2}\np3: {p3}\n\np: {p}')
```

```python
def Binomial(n,p,k):
    C_n_k = f(n)/(f(k)*f(n-k))
    return C_n_k * p**k * (1-p)**(n-k)    
```

```python
# Вероятность того, что ровно 3 мяча - белые
# Решение через Биномиальное уравнение:
C1 = Binomial(2,5/8,2) * Binomial(4,5/12,1) # (2:1)
C2 = Binomial(2,5/8,1) * Binomial(4,5/12,2) # (1:2)
C3 = Binomial(2,5/8,0) * Binomial(4,5/12,3) # (0:3) 
P = C1 + C2 + C3
P
```

##### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.


```python
# Решение через формулу Байеса
p1 = 0.9
p2 = 0.8
p3 = 0.6
P_A = 1 / 3 * (p1 + p2 + p3) # Полная вероятность

P1 = 1/3* p1 / P_A
P2 = 1/3* p2 / P_A
P3 = 1/3* p3 / P_A
print(f'P1: {P1}\nP2: {P2}\nP3: {P3}')
```

```python
print(f'P1: {p1/(p1+p2+p3)}\nP2: {p2/(p1+p2+p3)}\nP3: {p3/(p1+p2+p3)}')
```

##### 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
p1 = 0.8
p2 = 0.7
p3 = 0.9

P_A = 1/4*p1 + 1/4*p2 + 2/4*p3 # Полная вероятность
P1 = 1/4* p1 / P_A
P2 = 1/4* p2 / P_A
P3 = 2/4* p3 / P_A
print(f'P1: {P1}\nP2: {P2}\nP3: {P3}')
```

##### 5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
p1, q1 = 0.1, 1-0.1
p2, q2 = 0.2, 1-0.2
p3, q3 = 0.25, 1-0.25

P1 = p1 * p2 * p3 # выйдут из строя все детали (произошли все 3 события)
P2 = (p1 * p2 * q3) + (p1 * q2 * p3) + (q1 * p2 * p3) # выйдет из строя две детали
P3 = (p1 * q2 * q3) + (q1 * p2 * q3) + (q1 * q2 * p3) # выйдет из строя хотя бы одна деталь
P4 = 1 - (P1 + (q1 * q2 * q3))  # выйдут из строя от 1 до 2-х деталей (P2+P3) (противоположно собития отказа 0,3 деталей)

print(f'''    P1 (выйдут из строя все детали): {round(P1,3)}
    P2 (выйдут из строя только две детали): {round(P2,3)}
    P3 (выйдет из строя хотя бы одна деталь): {round(P3,3)} 
    P4 (выйдут из строя от 1 до 2-х деталей): {round(P4,3)}''')
```
